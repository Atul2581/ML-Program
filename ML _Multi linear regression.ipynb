```python
x=[]
y=[]
for line in lines:
    w = line.strip().split(";")
    ins = [float(v) for v in w[0:-1]]
    lab= float(w[-1])
    x.append(ins)
    y.append(lab)

import numpy as np
x = np.array(x)
ones = np.ones(x.shape[0])
X = np.c_[ones,x]
Y = np.c_[y]
np.random.seed(1)
W = 2*np.random.random((X.shape[1],1))-1
#print(W.shape)

def predict(x,w):
    return x.dot(w)

def loss(y,ycap):
    return((y-ycap)**2).mean()

def derivative(x,y,w):
    ycap = predict(x,w)
    return (x.T.dot(y-ycap))/x.shape[0]

ycap1 = predict(X,W)
print('loss:',loss(Y,ycap1))

grad = derivative(X,Y,W)
print(grad)
W+= grad*0.002
ycap2 = predict(X,W)
print('loss:',loss(Y,ycap2))
#scaling
def sd(x):
    return(((x-x.mean())**2).sum()/(x.size-1))**0.5

def scale(x):
    return (x-x.mean())/sd(x)

def scaleMatrix(x):
    for i in range(x.shape[1]):
        cal =x[:,i]
        x[:,i]=scale(cal)
    o = np.ones(x.shape[0])
    return np.c_[o,x]
    
```

```python
def train(x,y,w,alpha,iter,conv=0.000000001):
    ploss=0
    flag=0
    for i in range(iter):
        ycap = predict(x,w)
        closs = loss(y,ycap)
        print(closs)
        if i% 1000==0:
            print("loss at itration",i,"is",closs)
        diff = abs(ploss-closs)
        if diff<=conv:
            print("training completed",i+1,"iteration")
            flag=1
            break
        grad = derivative(x,y,w)
        w+= grad*alpha
        ploss=closs
    if flag==0:
        print("traninig not completed")
    return w
        
    
```

```python
theta = np.array(W)
print(theta)
theta = train(X,YL,theta,0.02,1000)
theta = train(X,YL,theta,0.02,8000)
```

```python
def accu(y,ycap,cloesness):
    ds = 100-cloesness
    dist = abs(y-ycap)/abs(y)*100
    pcnt = dist[dist<=ds].size
    n = y.shape[0]
    return pcnt/n*100
 
```

```python
# Find the min and max values for each column
def dataset_minmax(dataset):
    minmax = list()
    for i in range(len(dataset[0])):
        col_values = [row[i] for row in dataset]
        value_min = min(col_values)
        value_max = max(col_values)
        minmax.append([value_min, value_max])
    return minmax
    
```
